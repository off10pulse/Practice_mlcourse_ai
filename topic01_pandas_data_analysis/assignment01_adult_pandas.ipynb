```python
import pandas as pd
```

```python
# data = pd.read_csv("../../data/adult.data.csv")
data = pd.read_csv('https://github.com/Yorko/mlcourse.ai/raw/main/data/adult.data.csv')
data.head()
```

**1. Сколько мужчин и женщин (признак *sex*) представлено в этом наборе данных?**

```python
# Ваш код здесь
data['sex'].shape[0]
data[data['sex'] == 'Female'].shape[0]

# data['sex'].value_counts()
```

**2. Каков средний возраст (признак *age*) женщин?**

```python
# Ваш код здесь
data[data['sex'] == 'Female']['age'].mean()
```

**3. Какова доля граждан Германии (признак *native-country*)?**

```python
# Ваш код здесь
data[data['native-country'] == 'Germany'].shape[0] / data.shape[0] * 100
```

**4-5. Каковы средние значения и среднеквадратичные отклонения возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год? **

```python
# Ваш код здесь
print(data[data['salary'] == '>50K']['age'].mean(), data[data['salary'] == '>50K']['age'].std())
print(data[data['salary'] == '<=50K']['age'].mean(), data[data['salary'] == '<=50K']['age'].std())
```

**6. Правда ли, что люди, которые получают больше 50k, имеют как минимум высшее образование? (признак *education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* или *Doctorate*)**

```python
# Ваш код здесь
data[data['salary'] == '>50K']['education'].value_counts()
```

**7. Выведите статистику возраста для каждой расы (признак *race*) и каждого пола. Используйте *groupby* и *describe*. Найдите таким образом максимальный возраст мужчин расы *Amer-Indian-Eskimo*.**

```python
# Ваш код здесь
data[['age', 'race']].groupby(['race']).describe()
data[['age', 'sex']].groupby(['sex']).describe()
```

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
# Ваш код здесь
import numpy as np

total = np.array(data[data['salary'] == '>50K']['marital-status'])
total = np.array(list(map(lambda x: x if 'Married' in x else pd.NaT, total)))

new_ser = pd.Series(total)
married = new_ser.dropna().shape[0]
non_married = len(total) - married
print(married / len(total) * 100, non_married / len(total) * 100)

# data[(data["sex"] == "Male")
#      & (data["marital-status"].str.startswith("Married"))][
#     "salary"
# ].value_counts(normalize=True)

# data[
#     (data["sex"] == "Male")
#     & ~(data["marital-status"].str.startswith("Married"))
# ]["salary"].value_counts(normalize=True)
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
# Ваш код здесь
max_hours = data['hours-per-week'].max()
trudogolics = data[data['hours-per-week'] == 99].shape[0]
real_trudogolics = data[(data['hours-per-week'] == 99) & (data['salary'] == '>50K')].shape[0] / trudogolics * 100
print(max_hours, trudogolics, real_trudogolics)
```

**10. Посчитайте среднее время работы (*hours-per-week*) зарабатывающих мало и много (*salary*) для каждой страны (*native-country*).**

```python
# Ваш код здесь
last = data[['hours-per-week', 'native-country', 'salary']]
less_money = last[last['salary'] == '<=50K'].drop(['salary'], axis=1)
more_money = last[last['salary'] == '>50K'].drop(['salary'], axis=1)
less_money.groupby(['native-country']).mean(), more_money.groupby(['native-country']).mean()

# for (country, salary), sub_df in data.groupby(["native-country", "salary"]):
#     print(country, salary, round(sub_df["hours-per-week"].mean(), 2))

# pd.crosstab(
#     data["native-country"],
#     data["salary"],
#     values=data["hours-per-week"],
#     aggfunc=np.mean,
# ).T
```
